```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
w1 = np.random.rand()
w2 = np.random.rand()
w3 = np.random.rand()
w4 = np.random.rand()
w5 = np.random.rand()
w6 = np.random.rand()
b = np.random.rand()

```

```python
WineData = pd.read_csv('Training.csv')
```

```python
def sigmoid(x):
    return (1/(1+np.exp(-x)))
def sigmoid_prime(x):
    return sigmoid(x) * (1-sigmoid(x))
```

```python
Data = list()
outputData = list()
outputData = WineData['Wine'].tolist()
for i in range(len(outputData)):
    if outputData[i] == 2:
        outputData[i] = 0
Data.append(outputData)
Data.append(WineData['Alcohol'].tolist())
Data.append(WineData['Acl'].tolist())
Data.append(WineData['Phenols'].tolist())
Data.append(WineData['Flavanoids'].tolist())
Data.append(WineData['Proanth'].tolist())
Data.append(WineData['Proline'].tolist())
    
```

```python
Wine_Data = pd.DataFrame(
    {'Wine': Data[0],
     'Alcohol': Data[1],
     'Acl': Data[2],
     'Phenols': Data[3],
     'Flavanoids': Data[4],
     'Proanth': Data[5],
     'Proline': Data[6]   
    })
```

```python
learning_rate = 0.01
```

```python
for i in range(100000):
    ri = np.random.randint(len(Data[0]))
    point = Wine_Data.loc[[ri]]
    
    v1 = float(point['Alcohol'])
    v2 = float(point['Acl'])
    v3 = float(point['Phenols'])
    v4 = float(point['Flavanoids'])
    v5 = float(point['Proanth'])
    v6 = float(point['Proline'])
    
    
      
    z = v1 * w1 + v2 * w2 + v3 * w3 + v4 * w4 + v5 * w5 + v6 * w6 + b
#     print(float(point['Alcohol']))
#     print(float(point['Acl']))
#     print(float(point['Phenols']))
#     print(float(point['Flavanoids']))
#     print(float(point['Proanth']))
#     print(float(point['Proline']))
    pred =  sigmoid((z-100)/100)
    
  
    target = float(point['Wine'])
    cost = np.square(pred - target)
    
    if(i % 1000 == 0):
        print(cost)
    
    dcost_pred = 2 * (pred - target)
    dpred_dz = sigmoid_prime(z)
   
    dz_db = 1
    
    dcost_dz = dcost_pred * dpred_dz
    
    dcost_dw1 = dcost_pred * v1
    dcost_dw2 = dcost_pred  * v2
    dcost_dw3 = dcost_pred  * v3
    dcost_dw4 = dcost_pred  * v4
    dcost_dw5 = dcost_pred  * v5
    dcost_dw6 = dcost_pred * v6
    dcost_db = dcost_pred * dz_db
    
    
    
    w1 = w1 - learning_rate * dcost_dw1
    w2 = w2 - learning_rate * dcost_dw2
    w3 = w3 - learning_rate * dcost_dw3
    w4 = w4 - learning_rate * dcost_dw4
    w5 = w5 - learning_rate * dcost_dw5
    w6 = w6 - learning_rate * dcost_dw6
    
    
    
```

```python
Validation = pd.read_csv('Validation.csv')
Vali_data = pd.DataFrame(Validation)
length = len(Vali_data['Wine'])

for i in range(length):
    rand = np.random.randint(length)
    point = Vali_data.loc[[rand]]
    v1 = float(point['Alcohol'])
    v2 = float(point['Acl'])
    v3 = float(point['Phenols'])
    v4 = float(point['Flavanoids'])
    v5 = float(point['Proanth'])
    v6 = float(point['Proline'])
    z = b + w1*v1 + w2*v2+ w3*v3+ w4*v4+ w5*v5+ w6*v6
    if(sigmoid(z) > 0.5):
        print("Wine 1 " )
    else:
        print("Wine 2 ")
        
    
    
    
```
